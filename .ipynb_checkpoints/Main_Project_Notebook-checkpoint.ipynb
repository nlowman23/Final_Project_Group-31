## The Best Neighborhood
To determine the real best neighborhood, we decided to combine all 3 of our data sets into one big graph that contains the most important data from all of them. This will get us the neighborhood with the overall highest quality of public services available to its residents.

The code below makes a weighted graph with all of our metrics. This graph is made by calculating the proportion of its metric it fills up, and then combining it into one graph. IMPORTANT: Only neighborhoods that have data from ALL 3 data sets are included, since it wouldn't make sense for a neighborhood that we call the best to not have all 3 of these things. Here is how everything is weighted:
- Amount of facilities: 20%
- Amount of important facilities: 20%
- Education Score: 40%
- Sqft of parks: 20%

```python
# Setup styuff

import pandas as pd
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
import geopandas
data = pd.read_csv("https://data.wprdc.org/datastore/dump/fbb50b02-2879-47cd-abea-ae697ec05170")
df = pd.read_csv("Pittsburgh Education 2024.csv")
parks_area = pd.read_csv("ParkData2.csv")
parks_info = pd.read_csv("parkdata.csv")
neighborhoods = geopandas.read_file("Neighborhoods/Neighborhoods_.shp")
```

```python
# Loading data sets

# Jack A
facilities = data["neighborhood"].value_counts().sort_values(ascending=False) 
facilities_weighted = 0.2 * facilities / 412

# Jack B
least_important = ["OFFICE", "Office", "Salt Dome", "TRAINING", "Training", "Pool Closed", "Vacant", "Storage", "STORAGE", 
                   "SERVICE", "Service", "Concession", "Dugout", "Shelter", "Cabin", "Restrooms", "Utility"]
data_filtered = data[~data["type"].isin(least_important)]
data_filtered_grouped = data_filtered["neighborhood"].value_counts().sort_values(ascending=False)
important_facilities = 0.2 * (data_filtered_grouped / 132)

# lenny
school_counts = df['Neighborhood'].value_counts()
valid_neighborhoods = school_counts[school_counts >= 2].index
stability = df[df['Neighborhood'].isin(valid_neighborhoods)].groupby('Neighborhood')['Student Stability Rate'].mean()
approval = df[df['Neighborhood'].isin(valid_neighborhoods)].groupby('Neighborhood')['Teacher Approval Rate'].mean()
ratio_inverse = 100 - df[df['Neighborhood'].isin(valid_neighborhoods)].groupby('Neighborhood')['Student/Teacher Ratio'].mean()
combined = pd.concat([stability, approval, ratio_inverse], axis=1)
combined.columns = ['Stability', 'Approval', '100 - Ratio']
combined['Best Education Score'] = combined.mean(axis=1).sort_index()
bars = combined['Best Education Score'].sort_index().plot(kind='bar', color='orange', edgecolor='black')
combined['Best Education Score'].sort_values().cumsum()
# total score:  1041.846150
combined['Weighted Education Score'] = 0.4 * combined['Best Education Score'] / 1041.846150
education_weighted = combined.drop(['Best Education Score', 'Stability','Approval','100 - Ratio'],axis=1)
education_weighted = education_weighted.reset_index()# = 'neighborhood'
education_weighted = education_weighted.rename(columns={'Neighborhood': 'neighborhood'})
education_weighted = education_weighted.set_index('neighborhood')['Weighted Education Score']
#education_weighted.head()

# Nate
area = parks_area[['updatepknm', 'sqft']].copy() #keep only updated park name and sqr footage columns
area.rename(columns={'updatepknm': 'park_name'}, inplace=True) #rename the updated park name column so both files have the same name
info = parks_info[['name', 'neighborhood']].copy() #keep only the park name and neighborhood columns
info.rename(columns={'name': 'park_name'}, inplace=True) #rename the updated park name column so both files have the same name
area['park_name'] = area['park_name'].astype(str).str.strip().str.lower() #cleaning up the names so they match (made them lowercase and took away spaces)
info['park_name'] = info['park_name'].astype(str).str.strip().str.lower()
area['sqft'] = pd.to_numeric(area['sqft'], errors='coerce') #make the sqr footage numeric
merged = pd.merge(area, info, on='park_name', how='inner') #merge the two datasets with the cleaned up park names
neigh_totals = (merged.groupby('neighborhood', as_index=False)['sqft'].sum().sort_values('sqft', ascending=False)) #group the data by the neighborhood and add the sqr footage
#print(neigh_totals['sqft'].sum())
# 119731871.09870426 | total sqft
park_weighted = neigh_totals.set_index('neighborhood')['sqft']
park_weighted = 0.2 * park_weighted / 119731871.09870426
#park_weighted_mask = park_weighted > 10000
#park_weighted = park_weighted[park_weighted_mask]
#print(park_weighted.sort_values(ascending=False).head())
```

```python
# Graphing 
# for stacked bar graph  
merged = pd.merge(important_facilities, facilities_weighted, on='neighborhood', how='outer', suffixes=('_important', '_normal')).fillna(0)
merged = pd.merge(merged, park_weighted, left_on='neighborhood',right_on='neighborhood').fillna(0)
merged = pd.merge(merged, education_weighted, left_on='neighborhood',right_on='neighborhood').fillna(0)
merged['total'] = merged['count_important'] + merged['count_normal'] + merged['Weighted Education Score'] + merged['sqft']
merged = merged.sort_values('total', ascending=False)
#print(merged['total'].cumsum())
merged_graphthis = merged.drop('total', axis=1)
merged_graphthis.columns = ['# of Important Facilities', '# of Total Facilites', 'Amount of Park SQFT', 'Weighted Education Score']
merged_graphthis.plot(kind='bar',stacked=True,ylabel='Proportion',title='Proportions by Data Set (Stacked)')
plt.show()
# .46 + .37
```

Based on our metrics, we have determined that Brookline is the best neighborhood in Pittsburgh. It has a good education score, the 2nd largest amount of important facilities, and is 4th on public facilities overall. It also is 4th in total park area. Overall, Brookline is the jack of all trades we were looking for. Not only does it include lots of data from all 3 data sets, it also well exceeds the other neighborhoods in terms of stuff that is there. It has a good education, a large park area, and a bunch of public facilities, with many of them being important. 

While it may look like education score was weighted a lot more than the other categories, the graph only contained neighborhoods that had data from ALL 3 categories. We made this choice because we think that the best neighborhood has to offer everything to its residents. Since there were few neighborhoods that contained more than one school, the education scores will be a lot more concetrated than the other metrics which have a lot more data and a bigger spread. Even if education score was altered, Brookline would still be first. Here is what the graph would look like if there was minimal education score (note, proportions add up to less than 1.0 now, this graph is just a demonstration):

```python
combined['Weighted Education Score'] = 0.05 * combined['Best Education Score'] / 1041.846150
combinedNew = combined
education_weightedNew = combinedNew.drop(['Best Education Score', 'Stability','Approval','100 - Ratio'],axis=1)
education_weightedNew = education_weightedNew.reset_index()
education_weightedNew = education_weightedNew.rename(columns={'Neighborhood': 'neighborhood'})
education_weightedNew = education_weightedNew.set_index('neighborhood')['Weighted Education Score']
mergedNewEdu = pd.merge(important_facilities, facilities_weighted, on='neighborhood', how='outer', suffixes=('_important', '_normal')).fillna(0)
mergedNewEdu = pd.merge(mergedNewEdu, park_weighted, left_on='neighborhood',right_on='neighborhood').fillna(0)
mergedNewEdu = pd.merge(mergedNewEdu, education_weighted, left_on='neighborhood',right_on='neighborhood').fillna(0)
mergedNewEdu['total'] = mergedNewEdu['count_important'] + mergedNewEdu['count_normal'] + mergedNewEdu['Weighted Education Score'] + mergedNewEdu['sqft']
mergedNewEdu = mergedNewEdu.sort_values('total', ascending=False)
mergedNew_graphthis = mergedNewEdu.drop('total', axis=1)
mergedNew_graphthis.columns = ['# of Important Facilities', '# of Total Facilites', 'Amount of Park SQFT', 'Weighted Education Score']
mergedNew_graphthis.plot(kind='bar',stacked=True,ylabel='Proportion',title='Altered Proportions by Data Set (Stacked)')
plt.show()
```

As you can see, even if education score was weighted a lot less, Brookline is still a very obvious winner.

There's one more data set we'd like to show present: the heatmap of all the neighborhoods with facilities, parks, and schools.

```python
heatmap_data = merged['total']
heatmap_data.name = "hmap_data"
data_map = neighborhoods.merge(heatmap_data, how='left', left_on='hood', right_on='neighborhood')
pd.set_option('display.max_columns', None)
plotted_map = data_map.plot(column='hmap_data', 
               cmap="Blues",              
               edgecolor="black",        
               legend=True,              
               legend_kwds={'label': "Overall Proportion Score"}, 
               figsize=(10, 5),         
               missing_kwds={"color": "gray"}
               )


neighborhood_list = {"Brookline","Carrick","Point Breeze","Allegheny Center","East Liberty","Larimer","Beechview"}

for idx, row in neighborhoods.iterrows():
        if row["hood"] in neighborhood_list:
            centroid = row.geometry.centroid
            label = row["hood"]
            plotted_map.annotate(text=label, 
                    xy=(centroid.x, centroid.y), 
                    horizontalalignment='center', 
                    color='Black', 
                    fontsize=6.7
                    )
plt.show()
```

Here is where our neighborhoods lie. We thought it was interesting that there were two clumps of neighborhoods. Other than the clumps, however, there wasn't much of a pattern within the data. All neighborhoods in the heatmap are of pretty different sizes, and even though there are clumps the neighborhoods are pretty spread out. Even though there are clumps, the heatmap helps express the accuracy of our data and our conclusion because of the variety of locations and sizes of the neighborhoods we found to be top contenders. Therefore, beyond a shadow of a doubt, Brookline is the best neighborhood in Pittsburgh based on the quality of public services that the neighborhood provides to its residents.
