I am measuring the density of parks in each neighborhood of Pittsburgh. I am accessing 2 data sets for this. One of the data sets has each park linked to it's neighborhood, the other has each park linked to total size in square footage. My plan is to combine the two data sets into one graph that will show the number of parks in each neighborhood while taking into account a weight I have created using the square footage of the park from the 2nd data set as a metric. To combine these I will match the names of the parks together and put them under the neighborhood with their added weight value.

```python
import pandas as pd
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
df = pd.read_csv('parkdata.csv') #load dataset
parks_per_neighborhood = df['neighborhood'].value_counts().sort_values(ascending=False) #count parks per neighborhood
plt.figure(figsize=(12, 6)) #create the figure for the bar chart
parks_per_neighborhood.plot(kind='bar', color='skyblue', edgecolor='black') #draw bar graph
plt.title('Number of Parks in Each Neighborhood in Pittsburgh:', fontsize=16) #add a title to the chart
plt.xlabel('Neighborhood', fontsize=12) #x axis label
plt.ylabel('Number of Parks', fontsize=12) #y axis label
plt.tight_layout() #adjust the layout to better fit the text
plt.show() #show the graph
```

```python
import pandas as pd
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
df = pd.read_csv('parkdata.csv') #load dataset
topN = 10 #to show the top 10 neighborhoods
top_parks = parks_per_neighborhood.head(topN) #count the parks in each neighborhood and pick the top 10
plt.figure(figsize=(12, 6)) #create the figure
top_parks.plot(kind='bar', color='skyblue', edgecolor='black') #draw bar graph
plt.title('Neighborhoods by Number of Parks in Pittsburgh', fontsize=16) #add a title to the chart
plt.xlabel('Neighborhood', fontsize=12) #x axis label
plt.ylabel('Number of Parks', fontsize=12) #y axis label
plt.tight_layout() #adjust the layout to better fit the text
plt.show() #show the graph
```

```python
import pandas as pd
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
df = pd.read_csv('ParkData2.csv') #load dataset
df['sqft'] = pd.to_numeric(df['sqft'], errors='coerce') #convert the sqr foot column to numeric so I can do calculations
df['origpkname'] = df['origpkname'].astype(str) #make sure park names are stored as strings
total_sqft = df['sqft'].sum() #add the total area of all the parks together
df['weight'] = df['sqft'] / total_sqft #make a new column showing parks proportion of the total sqr footage
df_sorted = df.sort_values(by='sqft', ascending=False) #sort the parks from largest to the smallest by sqr footage
plt.figure(figsize=(14, 6)) #create a bar chart
plt.bar(df_sorted['origpkname'], df_sorted['weight'], color='gold', edgecolor='black') #plot the park names with their respective weights
plt.title('Weight of Each Park Based on Square Footage', fontsize=16) #add a title for the chart
plt.xlabel('Park Name', fontsize=12) #label the x axis
plt.ylabel('Weight (Proportion of Total Square Footage)', fontsize=10) #label the y axis
plt.xticks(rotation=90) #rotating the label names to try and make them readable (i know they still are not)
plt.tight_layout() #adjusting the spacing to try and make them more visible
plt.show() #show the finished chart
```

```python
import pandas as pd
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
df = pd.read_csv('ParkData2.csv') #load dataset
df['sqft'] = pd.to_numeric(df['sqft'], errors='coerce') #convert the sqr foot column to numeric so I can do calculations
df['origpkname'] = df['origpkname'].astype(str) #make sure park names are stored as strings
total_sqft = df['sqft'].sum()  #add the total area of all the parks together
df['weight'] = df['sqft'] / total_sqft #make a new column showing parks proportion of the total sqr footage
df_sorted = df.sort_values(by='sqft', ascending=False).head(10)  #sort the parks from largest to the smallest by sqr footage but take only the top 10
plt.figure(figsize=(14, 6)) #create a bar chart
plt.bar(df_sorted['origpkname'], df_sorted['weight'], color='gold', edgecolor='black') #plot the park names with their respective weights
plt.title('Weight of Each Park Based on Square Footage', fontsize=16) #add a title for the chart
plt.xlabel('Park Name', fontsize=12) #label the x axis
plt.ylabel('Weight (Proportion of Total Square Footage)', fontsize=10) #label the y axis
plt.xticks(rotation=90) #rotating the label names to try and make them readable (i know they still are not)
plt.tight_layout() #adjusting the spacing to try and make them more visible
plt.show() #show the finished chart
```

```python
import pandas as pd
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
parks_area = pd.read_csv("ParkData2.csv") #load the first data file  
parks_info = pd.read_csv("parkdata.csv")  #load the second data file

area = parks_area[['updatepknm', 'sqft']].copy() #keep only updated park name and sqr footage columns
area.rename(columns={'updatepknm': 'park_name'}, inplace=True) #rename the updated park name column so both files have the same name

info = parks_info[['name', 'neighborhood']].copy() #keep only the park name and neighborhood columns
info.rename(columns={'name': 'park_name'}, inplace=True) #rename the updated park name column so both files have the same name

area['park_name'] = area['park_name'].astype(str).str.strip().str.lower() #cleaning up the names so they match (made them lowercase and took away spaces)
info['park_name'] = info['park_name'].astype(str).str.strip().str.lower()

area['sqft'] = pd.to_numeric(area['sqft'], errors='coerce') #make the sqr footage numeric

merged = pd.merge(area, info, on='park_name', how='inner') #merge the two datasets with the cleaned up park names

neigh_totals = (merged.groupby('neighborhood', as_index=False)['sqft'].sum().sort_values('sqft', ascending=False)) #group the data by the neighborhood and add the sqr footage

plt.figure(figsize=(16, 7)) #create the bar graph
plt.bar(neigh_totals['neighborhood'], neigh_totals['sqft'], edgecolor='black')
plt.title('Total Park Area by Neighborhood (sq ft)', fontsize=16) #title
plt.xlabel('Neighborhood', fontsize=12) #x axis label
plt.ylabel('Total Park Area (sq ft)', fontsize=12) #y axis label
plt.xticks(rotation=90) #rotate the x axis labels
plt.tight_layout() #fit the layout
plt.show() #plot the final graph
```

Here I combined the amount of parks using the total landmass of the parks per neighborhood in sqr footage.

```python
import pandas as pd
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
parks_area = pd.read_csv("ParkData2.csv") #load the first data file  
parks_info = pd.read_csv("parkdata.csv")  #load the second data file

area = parks_area[['updatepknm', 'sqft']].copy() #keep only updated park name and sqr footage columns
area.rename(columns={'updatepknm': 'park_name'}, inplace=True) #rename the updated park name column so both files have the same name

info = parks_info[['name', 'neighborhood']].copy() #keep only the park name and neighborhood columns
info.rename(columns={'name': 'park_name'}, inplace=True) #rename the updated park name column so both files have the same name

area['park_name'] = area['park_name'].astype(str).str.strip().str.lower() #cleaning up the names so they match (made them lowercase and took away spaces)
info['park_name'] = info['park_name'].astype(str).str.strip().str.lower()

area['sqft'] = pd.to_numeric(area['sqft'], errors='coerce') #make the sqr footage numeric

merged = pd.merge(area, info, on='park_name', how='inner') #merge the two datasets with the cleaned up park names

neigh_totals = (merged.groupby('neighborhood', as_index=False)['sqft'].sum().sort_values('sqft', ascending=False)) #group the data by the neighborhood and add the sqr footage

top10 = neigh_totals.head(10) #get only the top 10 neighborhoods with most park area

plt.figure(figsize=(16, 7)) #create the bar graph
plt.bar(top10['neighborhood'], top10['sqft'], edgecolor='black') #create the graph with only the top 10 neighborhoods with parks
plt.title('Total Park Area by Neighborhood (sq ft)', fontsize=16) #title
plt.xlabel('Neighborhood', fontsize=12) #x axis label
plt.ylabel('Total Park Area (sq ft)', fontsize=12) #y axis label
plt.xticks(rotation=90) #rotate the x axis labels
plt.tight_layout() #fit the layout
plt.show() #plot the final graph
```

This shows the top 10 neighborhoods for parks factoring in the total landmass of the parks as well as the quantity of parks.
